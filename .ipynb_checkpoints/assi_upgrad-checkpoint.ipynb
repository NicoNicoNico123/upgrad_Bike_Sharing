# Upgrad Bike Sharing Assignment

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('day.csv')

# Display the first few rows of the DataFrame
df.info()
```

```python
df.sample(10)
```

## Visualize the data

```python
df_plot = df.drop(columns=['instant', 'dteday'])
sns.pairplot(df_plot)
```

## PLot the correlation to rental bike

```python
# Assuming your dataframe is loaded as df and target variable is 'cnt'
X = df.drop(['cnt','dteday','casual','registered','instant', 'yr'], axis=1)
y = df['cnt']

X.columns
```

```python
# Calculate the correlation of each feature with 'y'
correlations = X.corrwith(y)

plt.figure(figsize=(12, 6))
correlations.sort_values().plot(kind='bar', color='skyblue')
plt.title('Feature Correlation with Bike Rentals')
plt.xlabel('Features')
plt.ylabel('Correlation with cnt')
plt.axhline(0, color='red')
plt.show()
```

## Import RFE to select the feature out

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

model = LinearRegression()
rfe = RFE(estimator=model, n_features_to_select=5) # Select top 5 features
fit = rfe.fit(X, y)

# Plot feature ranking:
plt.figure(figsize=(12, 6))
plt.barh(X.columns, fit.ranking_)
plt.xlabel('Ranking')
plt.title('Feature Importance')
plt.grid(axis='x', linestyle='--')
plt.show()
```

## Import VIF to show any feature multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X_vif = add_constant(X)  # adding a constant column

vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(vif_data["Feature"], vif_data["VIF"], color='skyblue')
plt.xlabel('VIF')
plt.ylabel('Feature')
plt.title('VIF for each feature')
plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
```

### Let's drop more features and plot again

```python
X_sel = df.drop(['cnt','dteday','casual','registered','instant','temp','hum','windspeed','yr'], axis=1)
y_sel = df['cnt']

X_sel.columns
```

```python
X_vif = add_constant(X_sel)  # adding a constant column for intercept in the model

vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(vif_data["Feature"], vif_data["VIF"], color='lightgreen')
plt.xlabel('VIF')
plt.ylabel('Feature')
plt.title('VIF for each feature after drop some features')
plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
```

## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# Split X and y
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=42)
```

```python
X_train.columns
```

## Build model

```python
import statsmodels.api as sm

X_train_lm = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_lm).fit()

lr.params
```

```python
lr.summary()
```

```python
from sklearn.metrics import r2_score
# Add a constant to the test data
X_test_lm = sm.add_constant(X_test)

# Predict the values on the test dataset
y_test_pred = lr.predict(X_test_lm)

# Evaluate the model's performance on the test dataset
r2_model = r2_score(y_test, y_test_pred)
print (f" R2_score of model {r2_model}")
```
